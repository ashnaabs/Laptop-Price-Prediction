## Exploratory data analysis

```python
df1=df.drop(['Unnamed: 0'],axis=1)
```

```python
df1['Ram']=df1['Ram'].str.replace('GB','')
df1['Weight']=df1['Weight'].str.replace('kg','')
```

```python
df1['Ram']=df1['Ram'].astype('float32')
df1['Weight']=df1['Weight'].astype('float32')
```

```python
df1['ScreenResolution'].value_counts()
```

```python
df1['Touchscreen']=df1['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
```

```python
df1.head()
```

```python
df1['Ips']=df1['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
df1.head()
```

```python
new=df1['ScreenResolution'].str.split('x',n=1,expand=True)
```

```python
df1['x_res']=new[0]
df1['y_res']=new[1]
```

```python
df1['x_res']=df1['x_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
```

```python
df1.head()
```

```python
df1['x_res']=df1['x_res'].astype('int')
df1['y_res']=df1['y_res'].astype('int')
```

```python
df1['ppi']=(((df1['x_res']**2)+(df1['y_res']**2))**0.5/df1['Inches']).astype('float')
```

```python
df1.drop(columns=['ScreenResolution','x_res','y_res'],inplace=True)
```

```python
df1.head()
```

```python
df1['Cpu'].value_counts()
```

```python
df1['Cpu Name']=df1['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
df1.head()
```

```python
def fetch_processor(text):
  if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
      return text
  else:
      if text.split()[0] == 'Intel':
          return 'Other Intel Processor'
      else:
          return 'AMD processor'
```

```python
df1['cpu brand']=df1['Cpu Name'].apply(fetch_processor)
```

```python
df1.drop(columns=['Cpu','Cpu Name'],inplace=True)
```

```python
df1['cpu brand'].value_counts().plot(kind='bar')
```

```python
sns.barplot(x=df1['cpu brand'],y=df1['Price'])
plt.xticks(rotation='vertical')
```

```python
df1['Memory'].value_counts()
```

```python
df1['Memory']=df1['Memory'].astype(str).replace('\.0','',regex=True)
df1['Memory']=df1['Memory'].str.replace('GB','')
df1['Memory']=df1['Memory'].str.replace('TB','000')
new = df1['Memory'].str.split('+',n = 1,expand = True)

df1['first']=new[0]
df1['first']=df1['first'].str.strip()

df1['second']=new[1]


df1['layer1HDD']=df1['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df1['layer1SSD']=df1['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df1['layer1hybrid']=df1['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df1['layer1flash_storage']=df1['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

df1['second'].fillna('0',inplace=True)

df1['layer2HDD']=df1['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df1['layer2SSD']=df1['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df1['layer2hybrid']=df1['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df1['layer2flash_storage']=df1['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

df1['second']=df1['second'].str.replace(r'\D','')
df1['first']=df1['first'].str.replace(r'\D','')
df1['second']=df1['second'].str.strip()
df1['first']=df1['first'].str.replace('SSD','')
df1['second']=df1['second'].str.replace('SSD','')

df1['first']=df1['first'].astype(int)
df1['second']=df1['second'].astype(int)

```

```python
df1['HDD']=(df1['first']*df1['layer1HDD']+df1['second']*df1['layer2HDD'])
df1['SSD']=(df1['first']*df1['layer1SSD']+df1['second']*df1['layer2SSD'])
df1['Hybrid']=(df1['first']*df1['layer1hybrid']+df1['second']*df1['layer2hybrid'])
df1['Flash_Storage']=(df1['first']*df1['layer1flash_storage']+df1['second']*df1['layer2flash_storage'])
```

```python
df1.drop(columns=['first','second','layer1HDD','layer1SSD','layer1hybrid','layer1flash_storage','layer2HDD','layer2SSD','layer2hybrid','layer2flash_storage'],inplace=True)
```

# **Encoding**

```python
dfa=pd.get_dummies(df1[['Company','TypeName','cpu brand','Gpu brand','os']],drop_first=True)
dfa
```

```python
dfb=pd.concat([df1,dfa],axis=1)
dfb
```

```python
dfb = dfb.drop(['Company','TypeName','cpu brand','Gpu brand','os'], axis=1)
```

```python
dfb
```

```python
dfb.isna().sum()
```

### Filling missing values using mode function

```python
dfb['Inches']=dfb['Inches'].fillna(dfb['Inches'].mode()[0])
dfb['ppi']=dfb['ppi'].fillna(dfb['ppi'].mode()[0])
```

```python
x=dfb.drop(columns=['Price'])
y=np.log(dfb['Price'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=2)
```

# **Model** **creation** **and** **evaluation**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('r2 score is',r2_score(y_test,y_pred))
m1=r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
decision=DecisionTreeRegressor()
decision.fit(x_train,y_train)
y_pred1=decision.predict(x_test)
```

```python
print('r2 score is',r2_score(y_test,y_pred1))
m2=r2_score(y_test,y_pred1)
print('mape is',mean_absolute_percentage_error(y_test,y_pred1))
```

```python
from sklearn.ensemble import RandomForestRegressor
random=RandomForestRegressor()
random.fit(x_train,y_train)
y_pred2=random.predict(x_test)
```

```python
print('r2 score is',r2_score(y_test,y_pred2))
m3=r2_score(y_test,y_pred2)
print('mape is',mean_absolute_percentage_error(y_test,y_pred2))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1)
ridge.fit(x_train,y_train)
y_pred3=ridge.predict(x_test)
```

```python
print('r2 score is',r2_score(y_test,y_pred3))
m4=r2_score(y_test,y_pred3)
print('mape is',mean_absolute_percentage_error(y_test,y_pred3))
```

```python
import xgboost as xgb
xg_reg=xgb.XGBRegressor(n_estimators=100,max_depth=3)
xg_reg.fit(x_train,y_train)
y_pred4=xg_reg.predict(x_test)
```

```python
print('r2 score is',r2_score(y_test,y_pred4))
m5=r2_score(y_test,y_pred4)
print('mape is',mean_absolute_percentage_error(y_test,y_pred4))
```

## **Comparison** **of** **accuracy** **of** **models**

```python
models=['Linear_reg','Decision_tree','Random_forest','Ridge_reg','XGB']
accuracy=[m1,m2,m3,m4,m5]
plt.bar(models,accuracy,color=['blue','teal','black','pink','cadetblue'])
```
